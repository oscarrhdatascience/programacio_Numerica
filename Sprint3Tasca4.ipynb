- Exercici 3

Crea una funció que donada una taula de dues dimensions, et calculi els totals per fila i els totals per columna. 

```python
# Generate a function that generates a Matrix with random number of items between 2 and 5
def generateMatrix():
    numRows = np.random.randint (2,6)
    numColumns = np.random.randint (2,6)
    matrix2 = np.random.randint(1,101, size=(numRows, numColumns))
    return (matrix2)

# Generate a function that sums the row items and the column items 
def sumRowsSumColumns(matrix2):
    sumRows = np.sum(matrix2,1)
    sumColumns = np.sum(matrix2,0)
    return (sumRows, sumColumns)

# Main Program
matrix2 = generateMatrix()
print (matrix2)
sumRows, sumColumns = sumRowsSumColumns(matrix2)
print ("La suma de les files es =", sumRows)
print ("La suma de les columnes es =", sumColumns)
```

Implementa manualment una funció que calculi el coeficient de correlació. Informa’t-en sobre els seus usos i interpretació.

```python
import matplotlib.pyplot as plt

# Generate a function that first of all get N applying len to any of the arrays beause we will pass 2 arrays with the same size
# After getting N we calculate the mean of both arrays
# Then we apply the formula of covariance with our variables and now we have the numerator of the Correlation Coeficient
# To get the denominator we only have to calculate the standard deviation of both arrays and multiply them
# With the numerator and the denominator calculated we can finally return the Correlation Covariance
def coefCorrPear(arrX,arrY):
    #Size
    N=len(arrX)
    #Means
    meanX=np.mean(arrX)
    meanY=np.mean(arrY)
    #Covariance
    covXY=(sum(arrX * arrY) / N) - (meanX * meanY)
    #Standard deviation
    stdX=np.std(arrX)
    stdY=np.std(arrY)
    #Correlation Coeficient
    rXY = covXY / (stdX * stdY)
    return (rXY)

# In the Main Program we only generate an np.array with random numbers and a size value of 5, it's important that we get the same size on both arrays
# And then we call the function coefCorrPear
# Main Program
arrX= np.random.randint(1,100, size=(5))
arrY= np.random.randint(1,100, size=(5))
print (arrX)
print (arrY)
print ("El coeficient de correlació es = ",(coefCorrPear(arrX,arrY)))
print ("Al haber creat els arrays de forma aleatoria no solem tenir molta correlació entre un i l'altre")
plt.plot(arrX, arrY, marker = 'o')
plt.show()

print ("Si creem els arrays de forma voluntaria amb els mateixos valors la correlacio es total")
arrX2= np.array([1,2,3,4,5])
arrY2= np.array([1,2,3,4,5])
print (arrX2)
print (arrY2)
print ("El coeficient de correlació es = ",(coefCorrPear(arrX2,arrY2)))
plt.plot(arrX2, arrY2, marker = 'o')
plt.show()

print ("Si creem els arrays de forma voluntaria amb els valors invertits la correlacio es inversa")
arrX3= np.array([1,2,3,4,5])
arrY3= np.array([5,4,3,2,1])
print (arrX3)
print (arrY3)
print ("El coeficient de correlació es = ",(coefCorrPear(arrX3,arrY3)))
plt.plot(arrX3, arrY3, marker = 'o')
plt.show()
```
